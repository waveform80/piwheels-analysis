# PiWheels Usage

Now that PiWheels is quite mature, and we've got about 9 months worth of data let's revisit the usage patterns again. As before:

* Only actual downloads are logged in the table; accesses to directory indexes (either a result of `pip search` or the preliminary step in `pip install`) are not inserted.

* Any downloads which don't refer to extant entries in the `files` table aren't permitted.

* We can't yet tell where downloads are coming from: the `accessed_by` field is rather useless at present as it's just the Mythic Beasts IPv6 proxy address. This should be rectified after we migrate to the `piwheels.org` domain (we've now migrated but there may still be some users using the old address so we haven't by-passed the proxy *yet*).

* The `pip` user agent is dissected in the `downloads` table for easy analysis. We can check pip, python, Linux kernel, and distribution names and versions.

Let's get the imports and setup out of the way:

```python
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgres:///piwheels')
conn = engine.connect()
```

## Analysis

How many downloads are we dealing with and over what time period (excluding those rows where the Pi's clock wasn't set and hence read 1st Jan 1970)?

```python
print(conn.scalar(
    """
    SELECT COUNT(*) AS count
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    """))
```

```python
print(conn.scalar(
    """
    SELECT MAX(accessed_at) - MIN(accessed_at) AS duration
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    """))
```

How many downloads occurred each day?

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('day', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

The ramp up at the beginning probably represents the shift from Raspbian Jessie to Raspbian Stretch (which included the piwheels configuration by default).

It appears a steady state has now been reached (approx. from March 2018 onward). The large dips in May reflect a period during which the Pi's clock sync was having issues (subsequently resolved); there were downloads during those periods but they wound up being recorded against dates near the UNIX epoch (1970-01-01).

How many downloads are we handling per hour?

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > CURRENT_TIMESTAMP - INTERVAL '1 month'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

A daily cycle is fairly obvious from this graph. Let's zoom in to the latest week:

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > CURRENT_TIMESTAMP - INTERVAL '1 week'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

Is there a particular time of day that's popular?

```python
df = pd.read_sql(
    """
    SELECT
        TO_CHAR(accessed_at, 'HH24') AS hr,
        COUNT(*) AS downloads_per_hr
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    GROUP BY hr, EXTRACT(hour FROM accessed_at)
    ORDER BY EXTRACT(hour FROM accessed_at)
    """,
    conn, index_col='hr'
)
df.plot.bar(legend=False)
```

Yes, things pick up steadily from 07:00 onward (GMT) and peak around 15:00-20:00 (again GMT). Without geoip data (which we need to do some re-configuration to obtain) I'm not sure what to make of this; lots of UK based tinkerers in the evening? The big drop off from 9PM onward seems to suggest most piwheels users are based around European time-zones and head to bed at sensible hours!

Let's take a look at stats by day of week. Are most piwheels users weekend warriors?

```python
df = pd.read_sql(
    """
    SELECT
        TO_CHAR(accessed_at, 'Day') AS dow,
        COUNT(*) AS downloads_per_dow
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    GROUP BY dow, EXTRACT(isodow FROM accessed_at)
    ORDER BY EXTRACT(isodow FROM accessed_at) DESC
    """,
    conn, index_col='dow'
)
df.plot.barh(legend=False)
```

Yes! Saturday and Sunday seem consistently (slightly) busier. Friday seems slightly less busy (everyone heading down the pub early? ;)

How much time are we *really* saving users? In order to come up with a good estimate (while still erring on the side of caution) we need to take into consideration the following:

* People downloading pure Python wheels save little (if any) time. Unless the package contains a *huge* amount of Python the compilation time is minimal. We ought to disregard such wheels in our calculation.

* There's probably a certain minimum time that it takes to install *any* package whether from wheels or otherwise, largely dominated by the I/O time required to unpack it. We can probably calculate this from the median build time of pure Python packages (as the compilation time practically zero) and subtract it from any duration saved.

* Pi 2/3 users save some time, but Pi 0/1 users save *huge* amounts of time (because building is so slow on the single core platforms). We should try and estimate a rough multiplier for build times on the slower platforms and adjust saved time accordingly for armv6l downloads.

So here goes. Firstly, the median build time of pure Python packages. My guess is this is about 6 seconds.

```python
print(conn.scalar(
    """
    SELECT
        PERCENTILE_CONT(0.5)
            WITHIN GROUP (ORDER BY duration) AS median_dur
    FROM
        builds b
        JOIN files f ON b.build_id = f.build_id
    WHERE f.abi_tag = 'none'
    """,
    conn
))
```

Yup, ~6 seconds (for the sake of argument). After a bit of experimentation building numpy (which I'm hoping is vaguely representative of compilation times -- let me know if you've got more accurate figures!) I think a suitable multiplier for build durations to estimate time saved for Pi 0/1 users is 6x. That isn't too surprising given the Pi 2 when it came out was generally estimated to be roughly 6x the speed of the Pi B+, although I'm a little surprised it's so close (I was expecting less given that compilation of wheels doesn't *generally* involve parallel builds).

Anyway, taking all this into account, here's how much build time we've saved our users so far:

```python
print(conn.scalar(
    """
    SELECT
        SUM(
            CASE f.platform_tag
                WHEN 'linux_armv7l' THEN 1
                WHEN 'linux_armv6l' THEN 6
                ELSE 0
            END *
            CASE WHEN b.duration > INTERVAL '6.7 seconds'
                THEN b.duration - INTERVAL '6.7 seconds'
                ELSE INTERVAL '0'
            END
        ) AS time_saved
    FROM
        downloads d
        JOIN files f ON d.filename = f.filename
        JOIN builds b ON b.build_id = f.build_id
    WHERE f.abi_tag <> 'none'
    """))
```

From about 9 months of serving, we've saved users about *18 years* of build time!

How spread is this benefit? Evidently Pi 0/1 users save more time on an individual package but what about overall?

```python
df = pd.read_sql(
    """
    SELECT
        f.platform_tag,
        COUNT(*) AS downloads,
        SUM(
            CASE f.platform_tag
                WHEN 'linux_armv7l' THEN 1
                WHEN 'linux_armv6l' THEN 6
                ELSE 0
            END *
            CASE WHEN b.duration > INTERVAL '6.7 seconds'
                THEN b.duration - INTERVAL '6.7 seconds'
                ELSE INTERVAL '0'
            END
        ) AS time_saved
    FROM
        downloads d
        JOIN files f ON d.filename = f.filename
        JOIN builds b ON b.build_id = f.build_id
    WHERE f.abi_tag <> 'none'
    GROUP BY f.platform_tag
    """, conn, index_col='platform_tag')
df
```

So Pi 2/3 users are still benefitting most. I'd previously guessed that the 0s might overtake them simply due to the extra build time saved but apparently that's not the case; there's so many extra downloads from Pi 2 and 3 users that they're still easily in the lead.

Finally, let's take a quick look at the popularity of versions of Raspbian:

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('week', accessed_at) AS accessed,
        distro_version,
        COUNT(*) AS count
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    AND distro_name = 'Raspbian GNU/Linux'
    AND distro_version IN ('8', '9')
    GROUP BY accessed, distro_version
    ORDER BY distro_version, accessed
    """, conn)
df = df.pivot(
    index='accessed',
    columns='distro_version',
    values='count')
df.plot(logy=True)
```

It's curious to see Raspbian Jessie *increasing* in popularity, though I suspect this may simply be an increase in the overall popularity of the Pi itself. Still, it's several orders of magnitude lower than Stretch (note the logarithmic Y-axis).

How about the popularity of versions of CPython?

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('week', accessed_at) AS accessed,
        LEFT(py_version, 3) AS py_ver,
        COUNT(*) AS count
    FROM downloads
    WHERE accessed_at > '1970-01-05'
    AND py_name = 'CPython'
    AND distro_name = 'Raspbian GNU/Linux'
    AND distro_version IN ('8', '9')
    GROUP BY py_ver, accessed
    ORDER BY py_ver, accessed
    """, conn)
df = df.pivot(
    index='accessed',
    columns='py_ver',
    values='count')
df.plot(logy=True)
```

No surprise that 3.5 (the version that ships with Raspbian Stretch) is the most popular. However, it's intriguing that 2.7 is the next most popular, particularly given that we don't (yet) build binary wheels for that version. It's still an order of magnitude (again, log Y-axis) more popular than either 3.4 (the version shipped with Raspbian Jessie), an almost an order of magnitude more than 3.6.

Evidently our next efforts at supporting a version should probably be 2.7, despite its legacy nature.
