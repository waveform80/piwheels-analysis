# PiWheels Usage

Now that download logging has been properly integrated into PiWheels, and we've got a full months worth of data let's revisit the usage patterns. As before:

* Only actual downloads are logged in the table; accesses to directory indexes (either a result of `pip search` or the preliminary step in `pip install`) are not inserted.

* Any downloads which don't refer to extant entries in the `files` table aren't permitted (and furthermore deletions from the `files` table will cascade to remove entries from `downloads` too).

* We can't yet tell where downloads are coming from: the `accessed_by` field is rather useless at present as it's just the Mythic Beasts IPv6 proxy address. This should be rectified after we migrate to the `piwheels.org` domain.

* The `pip` user agent is dissected in the `downloads` table for easy analysis. We can check pip, python, Linux kernel, and distribution names and versions.

Let's get the imports and setup out of the way:

```python
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgres:///piwheels')
conn = engine.connect()
```

## Analysis

How many downloads are we dealing with and over what time period (excluding those rows where the Pi's clock wasn't set and hence read 1st Jan 1970)?

```python
print(conn.scalar("SELECT COUNT(*) AS count FROM downloads"))
```

```python
print(conn.scalar(
    """
    SELECT MAX(accessed_at) - MIN(accessed_at) AS duration
    FROM downloads
    WHERE accessed_at > '1970-01-02'::date
    """))
```

How many downloads occurred each day?

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('day', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > '1970-01-02'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

It looks like there's a steady increase in usage over time, which probably reflects a gradual shift from Raspbian Jessie to Raspbian Stretch in which the piwheels configuration is included by default.

How many downloads are we handling per hour?

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > '1970-01-02'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

A daily cycle is fairly obvious from this graph. Depending on where the peak is, we can take a guess at where most Pi users are located, geographically. First, let's look at a typical week (i.e. *not* over xmas!):

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at BETWEEN '2017-12-04' AND '2017-12-11'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

Monday to Friday look relatively similar with a peak fairly late in the day. Let's look at Wednesday:

```python
df = pd.read_sql(
    """
    SELECT
        DATE_TRUNC('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at BETWEEN '2017-12-06' AND '2017-12-07'
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False, figsize=(8, 4))
```

The peak appears around 20:00 GMT. Are most Pi users UK based and tinkering in the evenings? Or are most US based and using them during the day? (Or some combination?) We probably can't tell which until we sort out the proxy and get some geoip data but it looks like one of those two probabilities.

Saturday and Sunday seemed busier a couple of graphs back: lots of weekend Pi warriors! Is this a typical pattern?

```python
df = pd.read_sql(
    """
    SELECT
        TO_CHAR(accessed_at, 'Day') AS dow,
        COUNT(*) AS downloads_per_dow
    FROM downloads
    WHERE accessed_at BETWEEN '2017-12-04' AND '2017-12-11'
    GROUP BY dow, EXTRACT(isodow FROM accessed_at)
    ORDER BY EXTRACT(isodow FROM accessed_at) DESC
    """,
    conn, index_col='dow'
)
df.plot.barh(legend=False)
```

It would seem so. There's a bit of a spike on Tuesday (why?) but Saturday and Sunday seem consistently (slightly) busier.

How much time are we *really* saving users? In order to come up with a good estimate (while still erring on the side of caution) we need to take into consideration the following:

* People downloading pure Python wheels save little (if any) time. Unless the package contains a *huge* amount of Python the compilation time is minimal. We ought to disregard such wheels in our calculation.

* There's probably a certain minimum time that it takes to install *any* package whether from wheels or otherwise, largely dominated by the I/O time required to unpack it. We can probably calculate this from the median build time of pure Python packages (as the compilation time practically zero) and subtract it from any duration saved.

* Pi 2/3 users save some time, but Pi 0/1 users save *huge* amounts of time (because building is so slow on the single core platforms). We should try and estimate a rough multiplier for build times on the slower platforms and adjust saved time accordingly for armv6l downloads.

So here goes. Firstly, the median build time of pure Python packages. My guess is this is about 6 seconds.

```python
print(conn.scalar(
    """
    SELECT
        PERCENTILE_CONT(0.5)
            WITHIN GROUP (ORDER BY duration) AS median_dur
    FROM builds
    """,
    conn
))
```

Yup, 6.5 seconds (for the sake of argument). After a bit of experimentation building numpy (which I'm hoping is vaguely representative of compilation times -- let me know if you've got more accurate figures!) I think a suitable multiplier for build durations to estimate time saved for Pi 0/1 users is 6x. That isn't too surprising given the Pi 2 when it came out was generally estimated to be roughly 6x the speed of the Pi B+, although I'm a little surprised it's so close (I was expecting less given that compilation of wheels doesn't *generally* involve parallel builds).

Anyway, taking all this into account, here's how much build time we've saved our users so far:

```python
print(conn.scalar(
    """
    SELECT
        SUM(
            CASE f.platform_tag
                WHEN 'linux_armv7l' THEN 1
                WHEN 'linux_armv6l' THEN 6
                ELSE 0
            END *
            CASE WHEN b.duration > INTERVAL '6.5 seconds'
                THEN b.duration - INTERVAL '6.5 seconds'
                ELSE INTERVAL '0'
            END
        ) AS time_saved
    FROM
        downloads d
        JOIN files f ON d.filename = f.filename
        JOIN builds b ON b.build_id = f.build_id
    WHERE f.abi_tag <> 'none'
    """))
```

From 49 days of serving, we've saved a stonking 1 year and 9 months of build time for users. Not bad! How spread is this benefit? Evidently Pi 0/1 users save more time on an individual package but what about overall?

```python
df = pd.read_sql(
    """
    SELECT
        f.platform_tag,
        COUNT(*) AS downloads,
        SUM(
            CASE f.platform_tag
                WHEN 'linux_armv7l' THEN 1
                WHEN 'linux_armv6l' THEN 6
                ELSE 0
            END *
            CASE WHEN b.duration > INTERVAL '6.5 seconds'
                THEN b.duration - INTERVAL '6.5 seconds'
                ELSE INTERVAL '0'
            END
        ) AS time_saved
    FROM
        downloads d
        JOIN files f ON d.filename = f.filename
        JOIN builds b ON b.build_id = f.build_id
    WHERE f.abi_tag <> 'none'
    GROUP BY f.platform_tag
    """, conn, index_col='platform_tag')
df
```

So Pi 2/3 users are currently benefitting most though in time that may shift as more 0s start using Raspbian Stretch.
