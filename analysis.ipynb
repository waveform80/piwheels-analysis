# PiWheels Build Failure Analysis

## Pre-requisites and setup

Grab all the libraries we need and set matplotlib to use a nice style.

```python
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
from sqlalchemy import create_engine, text, MetaData
```

Connect to the database (change the SQLAlchemy DSN below if your database is named something other than "piwheels_old"), and have SQLAlchemy's MetaData scrape all the tables

```python
engine = create_engine('postgres:///piwheels_old')
conn = engine.connect()
meta = MetaData()
meta.bind = conn
meta.reflect(views=True)
```

Build some views of the errors to make analysis a bit easier. Be warned that the first view (actually a materialized view) called `errors` takes a *very* long time to build - be patient!

```python
if 'errors' not in meta.tables:
    # Create the errors table which looks for common errors in the
    # output of failed builds
    conn.execute(
    """
    CREATE TABLE errors AS (
        WITH logs AS (
            SELECT
                b.package,
                b.version,
                ROW_NUMBER() OVER (PARTITION BY b.package, b.version) AS line_num,
                line
            FROM
                builds AS b,
                regexp_split_to_table(b.output, E'\n') AS line
            WHERE NOT b.status
        ),
        errors AS (
            SELECT
                package,
                version,
                CASE
                    WHEN line ~ '^[^ ]+Error' THEN regexp_replace(line, '^([^ ]+Error).*', '\\1')
                    WHEN line ~ '^No matching distribution' THEN 'No release'
                    WHEN line ~ 'fatal error: .*\\.h: No such file or directory' THEN  'Missing headers'
                END AS error,
                line_num,
                line
            FROM logs
        )
        SELECT * FROM errors
        WHERE error IS NOT NULL
    )
    """
    )
    conn.execute(
    """
    ALTER TABLE errors
        ADD CONSTRAINT errors_pk PRIMARY KEY (package, version, line_num)
    """
    )
    conn.execute(
    """
    CREATE INDEX errors_error
        ON errors(error)
    """
    )

if 'last_errors' not in meta.tables:
    # Build the last_errors table which consists of the final error found
    # for each failed build
    conn.execute(
    """
    CREATE VIEW last_errors AS (
        SELECT
            e.package,
            e.version,
            e.error,
            e.line_num,
            e.line
        FROM
            errors AS e
            JOIN (
                SELECT
                    package,
                    version,
                    MAX(line_num) AS line_num
                FROM errors
                GROUP BY package, version
            ) AS l
                ON e.package = l.package
                AND e.version = l.version
                AND e.line_num = l.line_num
    )
    """
    )
```

## Analysis

Time to get on the with analysis. First let's look at the top ten reasons for build failure:

```python
df = pd.read_sql(
    """
    SELECT error, COUNT(*)
    FROM last_errors
    GROUP BY error
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='error'
)
```

```python
df[:10]
```

```python
df[:10].plot.barh(legend=False)
```

No surprise that "No release" is the major reason for build failure. What *is* interesting is that the next most common reasons are import failure (effectively missing dependencies) and syntax error, and that they're almost equal. As we're building all versions I expected there to be quite a bit of failure from old packages that were 3.x incompatible; frankly I'm a little surprised `SyntaxError` isn't higher.

Which are the most common packages we're missing?

```python
df = pd.read_sql(
    """
    SELECT
        regexp_replace(line, '.*(?:cannot import name|No module named) ''([^'']+)''.*', '\\1') AS module,
        COUNT(*)
    FROM last_errors
    WHERE error = 'ImportError'
    GROUP BY module
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='module'
)
```

```python
df[:20]
```

```python
df[:20].plot.barh(legend=False)
```

No surprises that `numpy` is the top of the list. `ConfigParser` being next is interesting because it indicates that a lot of the `ImportError` cases were saw are actually 2.x incompatibility errors. This explains why `SyntaxError` was so low.

There's some good candidates here for inclusion as dependencies in the next run. For example:

* numpy
* Cython
* pykern
* yaml

I'm not so convinced about installing django, turbogears, flask  and other web frameworks. I think many of the packages that depend on those will either be pure Python or be quite small and hence having wheels for them isn't so important but given they do seem to be quite numerous I could be persuaded otherwise?

In the meantime, what about missing headers...

```python
df = pd.read_sql(
    """
    SELECT
        regexp_replace(line, '.*fatal error: ([^:]+\\.h): .*', '\\1') AS module,
        COUNT(*)
    FROM last_errors
    WHERE error = 'Missing headers'
    GROUP BY module
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='module'
)
```

```python
df[:20]
```

```python
df[:20].plot.barh(legend=False)
```

Some interesting ones here. Particularly `ldap.h`, `couchbase.h`, `hdf5.h`. From a quick perusal of the list some obvious packages to install would be:

* zlib1g-dev
* libpng12-dev
* libffi-dev

## Fun stuff

How did the build rate vary?

```python
df = pd.read_sql(
    """
    SELECT
        date_trunc('hour', build_timestamp) AS built_at,
        COUNT(*) AS packages_per_hour
    FROM builds
    WHERE status
    GROUP BY built_at
    ORDER BY built_at
    """,
    conn, index_col='built_at'
)
df.plot(legend=False)
```

It's fairly obvious from the above when we scaled up to using multiple slaves. But what about the build-rate per slave? Did each slave manage as much running in parallel as the first one did?

```python
df = pd.read_sql(
    """
    SELECT
        date_trunc('hour', build_timestamp) AS built_at,
        built_by,
        COUNT(*) AS packages_per_hour
    FROM builds
    WHERE status
    GROUP BY built_at, built_by
    ORDER BY built_at, built_by
    """,
    conn
)
df.pivot('built_at', 'built_by').plot(legend=False)
```

More or less is the answer. The average build rate seems to be in the 500 packages per hour region, with the odd spike here and there (note that in the queries above we're only looking at successfully built packages as failed packages tend to fail quickly and would rather skew the build rate upwards).
