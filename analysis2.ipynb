# PiWheels (rev3) Build Failure Analysis

## Pre-requisites and setup

Grab all the libraries we need and set matplotlib to use a nice style.

```python
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
import numpy as np
from sqlalchemy import create_engine, text, MetaData
```

Connect to the database (change the SQLAlchemy DSN below if your database is named something other than "piwheels"), and have SQLAlchemy's MetaData scrape all the tables

```python
engine = create_engine('postgres:///piwheels')
conn = engine.connect()
meta = MetaData()
meta.bind = conn
meta.reflect(views=True)
```

Build some views of the errors to make analysis a bit easier. Be warned that the first view (actually a materialized view) called `errors` takes a *very* long time to build - be patient!

```python
if 'errors' not in meta.tables:
    # Create the errors table which looks for common errors in the
    # output of failed builds
    conn.execute(
    """
    CREATE TABLE errors AS (
        WITH logs AS (
            SELECT
                b.package,
                b.version,
                ROW_NUMBER() OVER (PARTITION BY b.package, b.version) AS line_num,
                line
            FROM
                builds AS b,
                regexp_split_to_table(b.output, E'\n') AS line
            WHERE NOT b.status
        ),
        errors AS (
            SELECT
                package,
                version,
                CASE
                    WHEN line ~ '^[^ ]+Error' THEN regexp_replace(line, '^([^ ]+Error).*', '\\1')
                    WHEN line ~ '^No matching distribution' THEN 'No release'
                    WHEN line ~ 'fatal error: .*\\.h: No such file or directory' THEN  'Missing headers'
                END AS error,
                line_num,
                line
            FROM logs
        )
        SELECT * FROM errors
        WHERE error IS NOT NULL
    )
    """
    )
    conn.execute(
    """
    ALTER TABLE errors
        ADD CONSTRAINT errors_pk PRIMARY KEY (package, version, line_num)
    """
    )
    conn.execute(
    """
    CREATE INDEX errors_error
        ON errors(error)
    """
    )

if 'last_errors' not in meta.tables:
    # Build the last_errors table which consists of the final error found
    # for each failed build
    conn.execute(
    """
    CREATE VIEW last_errors AS (
        SELECT
            e.package,
            e.version,
            e.error,
            e.line_num,
            e.line
        FROM
            errors AS e
            JOIN (
                SELECT
                    package,
                    version,
                    MAX(line_num) AS line_num
                FROM errors
                GROUP BY package, version
            ) AS l
                ON e.package = l.package
                AND e.version = l.version
                AND e.line_num = l.line_num
    )
    """
    )
```

## Analysis

Time to get on the with analysis. First let's look at the top ten reasons for build failure:

```python
df = pd.read_sql(
    """
    SELECT error, COUNT(*)
    FROM last_errors
    GROUP BY error
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='error'
)
df[:10]
```

```python
df[:10].plot.barh(legend=False)
```

This time around the major cause of failure is `SyntaxError` which in practice means 2.x-only compatibility. Let's have a look at `ImportError` in more depth.

```python
df = pd.read_sql(
    """
    SELECT
        regexp_replace(line, '.*(?:cannot import name|No module named) ''([^'']+)''.*', '\\1') AS module,
        COUNT(*)
    FROM last_errors
    WHERE error = 'ImportError'
    GROUP BY module
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='module'
)
df[:40]
```

```python
df[:20].plot.barh(legend=False)
```

A surprising number of these import errors are *also* down to 2.x-only compatibility. For example, the following are all modules / functions that were renamed in 3.x:

* ConfigParser
* walk
* urlparse
* urllib2
* StringIO (and cStringIO)
* httplib
* izip

Interestingly, the following modules are also all down to 2.x-only compatibility but of a different sort. These turn out to be "common" names for modules in packages. In many cases, developers then attempt to import them relatively (while 3.x enforces absolute imports):

* version
* core
* base
* utils
* main
* future
* common

Of the remaining set, turbogears was discontinued (in favour of pylons) and thus never transitioned to 3.x. The setupfiles dependency is genuine ... but the packages that depend on it are all tiny (almost one function) packages by the same author.

Which finally brings us to the dependencies we ought to concern ourselves with for future runs:

* pykern - this is a genuine dependency although there's not *many* packages that use it (just a lot of versions). Unfortunately it's horrifically complex to install due to its own dependencies (most of which are later than the equivalent Raspbian packages). Might be worth revisiting when stretch lands.
* networkx - we should just add this.
* fabric - needs further investigation but it's a candidate.
* git - ditto.
* watchdog - needs investigation to determine what uses it.
* demjson - ditto.
* pybind11 - ditto.
* cloudmesh_base - ditto.

Onto the missing headers, which now form a tiny minority of failed packages:

```python
df = pd.read_sql(
    """
    SELECT
        regexp_replace(line, '.*fatal error: ([^:]+\\.h): .*', '\\1') AS module,
        COUNT(*)
    FROM last_errors
    WHERE error = 'Missing headers'
    GROUP BY module
    ORDER BY COUNT(*) DESC
    """,
    conn, index_col='module'
)
df[:20]
```

```python
df[:20].plot.barh(legend=False)
```

Windows we can ignore for obvious reasons. libxml, lapacke, and hdf5 are definitely worth checking for future builds. Further investigation is required for the couchbase, pattern, lua, smbclient, and sqlite3 headers (why is sqlite3 even there given Python's had sqlite3 built in for ages?).

The failures related to the GL headers turn out to be almost entirely down to Kivy. However, this is an important enough package that we should check whether we can include it.

## Fun stuff

How did the build rate vary? Firstly a quick'n'crude summation of builds per hour:

```python
df = pd.read_sql(
    """
    SELECT
        date_trunc('hour', built_at) AS ts,
        COUNT(*) AS packages_per_hour
    FROM builds
    WHERE status
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
```

Now a slightly more nuanced moving average calculated with a sliding day-long window. This shows an initially high build-rate (achieved by a doubling up of slave instances per Pi) gave way to a steady state of ~3000 pkgs/hour (after the doubled slaves were abandoned due to a high rate of slaves crashing from out-of-memory errors):

```python
df = pd.read_sql(
"""
WITH build_times AS (
    SELECT
        date_trunc('hour', built_at) AS hour,
        COUNT(*) AS builds
    FROM builds
    GROUP BY hour
),
all_times AS (
    SELECT
        d.d AS hour
    FROM generate_series(
        (SELECT date_trunc('hour', MIN(built_at)) FROM builds),
        (SELECT date_trunc('hour', MAX(built_at)) FROM builds),
        INTERVAL '1 hour'
    ) AS d
)
SELECT
    a.hour,
    (SUM(COALESCE(builds, 0)) OVER (
        ORDER BY a.hour
        ROWS BETWEEN 24 PRECEDING AND CURRENT ROW
    )) / 24 AS builds
FROM
    all_times a
    LEFT JOIN build_times b ON a.hour = b.hour
ORDER BY hour
""", conn, index_col='hour')
df.plot.line(legend=False)
```

And a moving-average of builds per minute calculated with an hour-long sliding window. This shows how the build-rate *occassionally* drops to zero briefly when complex packages like numpy or Kivy are building:

```python
df = pd.read_sql(
"""
WITH build_times AS (
    SELECT
        date_trunc('minute', built_at) AS minute,
        COUNT(*) AS builds
    FROM builds
    GROUP BY minute
),
all_times AS (
    SELECT
        d.d AS minute
    FROM generate_series(
        (SELECT date_trunc('minute', MIN(built_at)) FROM builds),
        (SELECT date_trunc('minute', MAX(built_at)) FROM builds),
        INTERVAL '1 minute'
    ) AS d
)
SELECT
    a.minute,
    SUM(COALESCE(builds, 0)) OVER (
        ORDER BY a.minute
        ROWS BETWEEN 60 PRECEDING AND CURRENT ROW
    ) / 60 AS builds
FROM
    all_times a
    LEFT JOIN build_times b ON a.minute = b.minute
ORDER BY minute
""", conn, index_col='minute')
df.plot.line(legend=False)
```

Minimum, maximum and average build times. It's notable that the median and mean are significantly different so there's evidently quite a few outliers in the build times:

```python
df = pd.read_sql(
    """
    SELECT
        MIN(duration) AS shortest_build,
        MAX(duration) AS longest_build,
        PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY duration) AS median_build,
        AVG(duration) AS mean_build
    FROM builds
    WHERE status
    """,
    conn
)
df
```

Here's a histogram of the build times. As expected there's a fairly significant chunk of outliers at the top end (note the logarithmic Y-axis):

```python
df = pd.read_sql(
    """
    SELECT EXTRACT(EPOCH FROM duration) AS dur_seconds
    FROM builds
    WHERE status
    """,
    conn
)
df.plot.hist(bins=50, legend=False, logy=True)
```

Now a histogram of the durations of less than a minute. Note there's an interesting peak at 6 seconds. I'd guess that all builds involving compilation and linking probably have a minimum build time of about 6 seconds and everything below that is pure-Python packages:

```python
df = pd.read_sql(
    """
    SELECT EXTRACT(EPOCH FROM duration) AS dur_seconds
    FROM builds
    WHERE status
    AND duration < INTERVAL '1 minutes'
    """,
    conn
)
df.plot.hist(bins=50, legend=False, logy=True)
```

How about a histogram of file-sizes (again, note the logarithmic Y-axis)? This shows pretty much what we'd expect; the vast majority of builds are small with ever decreasing counts as the size increases:

```python
df = pd.read_sql(
    """
    SELECT filesize
    FROM files
    """,
    conn
)
df.plot.hist(bins=50, legend=False, logy=True)
```

How about the percentile distributions of these statistics? Firstly, the build times. Note again that strange "bump" at the start. It feels like two relatively normal distributions have been overlapped which is probably pure-Python vs compiled packages:

```python
df = pd.read_sql(
    """
    SELECT
        p.p * 100 AS percentile,
        percentile_disc(p.p) WITHIN GROUP (
            ORDER BY EXTRACT(EPOCH FROM duration)
        ) AS duration
    FROM builds, generate_series(0.0, 1.0, 0.02) AS p
    WHERE status
    GROUP BY p.p
    """,
    conn, index_col='percentile'
)
df.plot.bar(legend=False, logy=True, color='#ee4455', width=1.0)
```

And now file-sizes. This is about as "normal" as we'd expect from the prior chart, although it's telling that the "tail" is quite tall so the outliers are really huge by comparison to most files:

```python
df = pd.read_sql(
    """
    SELECT
        p.p * 100 AS percentile,
        percentile_disc(p.p) WITHIN GROUP (ORDER BY filesize) AS filesize
    FROM files, generate_series(0.0, 1.0, 0.02) AS p
    GROUP BY p.p
    """,
    conn, index_col='percentile'
)
df.plot.bar(legend=False, logy=True, color='#ee4455', width=1.0)
```

Is there any correlation between build time and package sizes? Not a great deal. Other than an expected cluster at the bottom left (short builds and small files), stuff is *fairly* randomly scattered:

```python
df = pd.read_sql(
    """
    SELECT
        f.filesize,
        EXTRACT(EPOCH FROM b.duration) AS duration
    FROM builds b JOIN files f ON b.build_id = f.build_id
    """,
    conn
)
df.plot.hexbin('duration', 'filesize', gridsize=25, bins='log')
```
