# PiWheels Usage

Time to look at how much PiWheels is being used, and what impact it's having. This is a thoroughly preliminary analysis; integration of Apache's logs into the database is not yet complete but I wanted to gauge how much traffic we're likely to be handling there and get a feel for how much space certain designs are going to require.

The design of the `downloads` table in the database *currently* looks like the following:

```sql
CREATE TABLE downloads (
    filename            VARCHAR(255) NOT NULL,
    accessed_by         INET NOT NULL,
    accessed_at         TIMESTAMP NOT NULL,
    user_data           JSONB NOT NULL,

    CONSTRAINT downloads_filename_fk FOREIGN KEY (filename)
        REFERENCES files (filename) ON DELETE CASCADE
);

CREATE INDEX downloads_files ON downloads(filename);
```

Some notes on the structure:

* Only actual downloads are logged in the table; accesses to directory indexes (either a result of `pip search` or the preliminary step in `pip install`) are not inserted.

* Any downloads which don't refer to extant entries in the `files` table aren't permitted (and furthermore deletions from the `files` table will cascade to remove entries from `downloads` too).

* The `accessed_by` field is rather useless at present as it's just the Mythic Beasts IPv6 proxy address. This should be rectified after we migrate to the `piwheels.org` domain.

* The JSON that `pip` reports in the User-Agent field is stored (more or less) verbatim in the `user_data` field for further analysis.

Anyway, on with the analysis. First let's get the imports out of the way.

````python
import io
import json
import gzip

%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
import lars.apache
from sqlalchemy import create_engine, text, MetaData
from sqlalchemy.exc import IntegrityError
from pathlib import Path
````

Next, connect to the database and, if necessary, construct the new table and load data from the log files into it:

````python
engine = create_engine('postgres:///piwheels')
conn = engine.connect()
meta = MetaData()
meta.bind = conn
meta.reflect(views=True)
````

````python
if 'downloads' not in meta.tables:
    with conn.begin():
        conn.execute("""
        CREATE TABLE downloads (
            filename            VARCHAR(255) NOT NULL,
            accessed_by         INET NOT NULL,
            accessed_at         TIMESTAMP NOT NULL,
            user_data           JSONB NOT NULL,

            CONSTRAINT downloads_filename_fk FOREIGN KEY (filename)
                REFERENCES files (filename) ON DELETE CASCADE
        )
        """)
        conn.execute("""
        CREATE INDEX downloads_files ON downloads(filename)
        """)
    meta.reflect(views=True)
    
    def is_download(row):
        return (
            row.status == 200
            and row.req_User_Agent is not None
            and row.req_User_Agent.startswith('pip/')
            and row.request.url.path_str.endswith('.whl')
        )

    logs = [
        p for p in Path('.').iterdir()
        if p.name.startswith('ssl_access.log')
    ]
    # Copied from the Apache config
    log_format = "%h %l %u %t \"%r\" %>s %O \"%{Referer}i\" \"%{User-Agent}i\""
    log_encoding = 'ascii'
    
    for log in logs:
        print('Processing', log)
        if log.suffix == '.gz':
            log_file = io.TextIOWrapper(gzip.open(str(log), 'rb'),
                                        encoding=log_encoding)
        else:
            log_file = io.open(str(log), 'r', encoding=log_encoding)
        source = lars.apache.ApacheSource(log_file, log_format=log_format)
        with conn.begin():
            for row in source:
                if is_download(row):
                    path = Path(row.request.url.path_str)
                    if '{' in row.req_User_Agent:
                        user_data = json.loads(
                            row.req_User_Agent[row.req_User_Agent.find('{'):])
                    else:
                        user_data = {}
                    try:
                        with conn.begin_nested():
                            conn.execute(
                                meta.tables['downloads'].insert(),
                                filename=path.name,
                                accessed_by=str(row.remote_host),
                                accessed_at=row.time,
                                user_data=user_data
                            )
                    except IntegrityError:
                        print('Skipped', path.name)
````

## Analysis

How many downloads are we dealing with and over what time period (excluding those rows where the Pi's clock wasn't set and hence read 1st Jan 1970)?

````python
pd.read_sql("SELECT COUNT(*) AS count FROM downloads", conn)
````

````python
pd.read_sql(
    """
    SELECT MAX(accessed_at) - MIN(accessed_at) AS duration
    FROM downloads
    WHERE accessed_at > '1970-01-02'::date
    """, conn)
````

How many downloads occurred each day?

````python
df = pd.read_sql(
    """
    SELECT
        date_trunc('day', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > '1970-01-02'::date
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
````

How many downloads per hour?

````python
df = pd.read_sql(
    """
    SELECT
        date_trunc('hour', accessed_at) AS ts,
        COUNT(*) AS downloads_per_hour
    FROM downloads
    WHERE accessed_at > '1970-01-02'::date
    GROUP BY ts
    ORDER BY ts
    """,
    conn, index_col='ts'
)
df.plot(legend=False)
````

The daily cycle is fairly obvious from this graph. Quite impressive that a single Pi3 is handling up to a thousand downloads an hour whilst coordinating all builds and hosting the PostgreSQL database!

How much time have we saved users, i.e. how long would it have taken, at a minimum, to build all the packages that have been downloaded? (note: this is an underestimate as it excludes all installations where a wheel downloaded from piwheels was already cached, and underestimates build times on smaller platforms like the Pi Zero)?

````python
pd.read_sql(
    """
    SELECT SUM(b.duration) AS total_build_time
    FROM
        builds b
        JOIN files f ON b.build_id = f.build_id
        JOIN downloads d ON f.filename = d.filename
    """, conn)
````

What are the most popular packages (my guess is things like numpy, matplotlib and pandas, but let's see)?

````python
df = pd.read_sql(
    """
    SELECT
        b.package,
        COUNT(*) AS downloads
    FROM
        builds b
        JOIN files f ON b.build_id = f.build_id
        JOIN downloads d ON f.filename = d.filename
    GROUP BY b.package
    ORDER BY downloads DESC
    """, conn, index_col='package')
df[:20]
````

````python
df[:20].plot.barh(legend=False)
````

Well, that's the first major surprise. I'm guessing there's some good reason "home-assistant-frontend" is *incredibly* popular (?!) but I don't know what it is off-hand.

What are the OS and Python versions that users are using?

````python
pd.read_sql(
    """
    SELECT
        user_data->'distro'->>'name' AS os,
        user_data->'distro'->>'version' AS os_version,
        user_data->>'cpu' AS arch,
        user_data->>'python' AS py_version,
        COUNT(*) AS downloads
    FROM downloads
    GROUP BY os, os_version, arch, py_version
    ORDER BY downloads DESC
    """, conn)
````

No huge surprise to see Raspbian Stretch, with Python 3.5.3 heading the list here. However, number 3 is quite a surprise! Apparently there's a whole load of Ubuntu Xenial users downloading packages for Python 2.7 from piwheels ...
